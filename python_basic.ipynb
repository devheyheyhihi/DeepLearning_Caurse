###덧셈뺄셈곱셈나눗셈

함수

```
def myAdd(a,b):
  return a+b
def myDel(a,b):
  return a-b
def myDup(a,b):
  return a*b
def myDiv(a,b):
  return a/b

  
a=int(input("a의 값 : "))
b=int(input("b의 값 : "))
print("myadd = ",myAdd(a,b))
print("myDel : {0} - {1} = {2}".format(a, b, myDel(a,b)))
```

클래스

```
class calc:
  def __init__(self,a,b):
    self.a=a
    self.b=b
  
  def Add(self):
    return self.a+self.b
  
  def Del(self):
    return self.a-self.b
  
  def Dup(self):
    return self.a*self.b
  
  def Div(self):
    return self.a/self.b

  def __del__(self):
    print ("exit")

a=10
b=4
temp = calc(a,b)
result = temp.Add()
print("{0}+{1}={2}".format(a,b, result))
result = temp.Del()
print("{0}-{1}={2}".format(a,b, result))
result = temp.Dup()
print("{0}*{1}={2}".format(a,b, result))
result = temp.Div()
print("{0}/{1}={2}".format(a,b, result))
```

###연산자 중복 정의


```
class GString:
  def __init__(self, init=None):
    self.content = init
  
  def __sub__(self, str): ##빼기 연산자를 중복
    for i in str:
      self.content = self.content.replace(i,'')
    return GString(self.content)

  def __add__(self, str): ##더하기 연산자 중복
    for i in str:
      self.content += i
    return GString(self.content)

  def __abs__(self):
    return GString(self.content.upper())

  def Print(self):
    print(self.content)



g = GString("aBcdef")
g-="df"
g.Print()

g=abs(g)
g.Print()

g+="add"
g.Print()
```
